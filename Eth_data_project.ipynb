```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# Calling API - blockchain.com (blockchain.info)

url = "https://api.blockchain.info/charts/transactions-per-second?timespan=1days&rollingAverage=1hours&format=json"  # Replace this with the URL you want to make a GET request to

# Send the GET request
response = requests.get(url)

# Check if the request was successful (status code 200 means success)
if response.status_code == 200:
    data = response.json()  # If the response contains JSON data
    
    pretty_json = json.dumps(data, indent=4)
    
    print(pretty_json)
else:
    print(f"Request failed with status code: {response.status_code}")
```

```python
# Process JSON file
arr_x = [pt["x"] for pt in data["values"]]
arr_y = [pt["y"] for pt in data["values"]]

plt.plot(arr_x, arr_y)
```

```python
# Replace this with your Unix timestamp
from_timestamp = arr_x[0]
to_timestamp = arr_x[-1]    

# Convert to a `datetime` object
from_dt = datetime.datetime.fromtimestamp(from_timestamp)
to_dt = datetime.datetime.fromtimestamp(to_timestamp)

print("From: ", from_dt)
print("To: ", to_dt)
```
